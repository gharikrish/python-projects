```python
""" THIS PROJECT IS HELPFUL FOR STUDENTS AND TEACHERS AND EDUCTION INSTITUTION TO CONDUCT AND PRACTICE MCQ QUESTIONS """



import os
import csv
questions = dict()
ans = list()
pw_file = [1111,'E:\Python Exercises\proj.csv'] # 1111 is password #next filepath and file
pwlist = [] # this for we know how many times we are changed password
mark = 0

#file reading(only csv)
with open(pw_file[1],'r') as file:
    aa = csv.reader(file)
    for i in aa:
        questions[i[0]] = i[1]
        ans.append(i[2:])
file.close()

def trainer():
    pw = int(input('enter the pw : '))
    if pw == pw_file[0]:
        print("1 --> RESET PASSWORD")
        print("2 --> CHANGE THE QUESTIONS " )
        opt = int(input('enter the opt : '))
        if  opt == 1:
            new_pw = int(input("enter the new password : "))
            pw_file.remove(pw_file[0])
            pw_file.insert(0,new_pw)
            pwlist.append(new_pw)
            print('successfully upload your password')
            
        elif opt == 2:
            file_and_question_path = str(input('enter the file path'))
            file_name = str(input('enter the file name'))
            join_file_and_path = file_and_question_path+'\\'+file_name
            print("1-->yes")
            op = int(input('enter the op'))
            if op == 1:
                pw_file.remove(pw_file[1])
                pw_file.append(join_file_and_path)
                print('successfully upload your file')
            else:
                print('you press the wrong')
        else:
            print('some thing wrong sorry for inconvience.')
                        

def student():
    print('Best of luck')
    global mark
    c = 0
    for i,j in questions.items():
        print(str(c+1)+".",i)
        print("A.",ans[c][0]," ","B.",ans[c][1]," ","C.",ans[c][2]," ","D.",ans[c][3])
        print("                                                                     ")
        c = c+1
        student_answer = str(input('enter the answer as per the font :'))
        if (i,student_answer) in questions.items() :
            mark = mark +1   
        else:
            continue     
                       
def MCQ(): # MCQ--> MULTI CHOICE QUESTION
    print("1 -->  trainer") 
    print("2 -->  student")
    choice = int(input("enter the choice : "))
    if choice == 1:
        trainer()
        
    elif choice == 2:
        student()
        print('             ')
        print('congratulations')
        print('your mark is ',mark)
    
    else:
        print("please enter the correct options")

        
        
# finally run the program        
MCQ()
            
```
